# Hierarchical Clustering Algorithm (Unsupervised)

```python
import pandas as pd
# DataFrame with dicsonary
data = pd.DataFrame(data = {'x': [0,1.1,1,2,2,4,5,5], 'y': [0,1.5,4,2,3,1,0,4]})
data
```

**Use Scatter plot**

```python
import matplotlib.pyplot as plt
plt.scatter(data.x, data.y)

# Naming data points
numbers = [0, 1, 2, 3, 4, 5, 6, 7]
for index, value in enumerate(numbers):
    plt.annotate(value, (data.x[index], data.y[index]), size=14)
```

### Dendrogram

```python
import scipy.cluster.hierarchy as sc
# Create method in scipy.cluster.hierarchy
dendrogram = sc.dendrogram(sc.linkage(data, method='ward'))
plt.title('Dendrogram') # plot Title
plt.ylabel('Euclidean distances') # plot Ylabel
plt.show()
```

### Train Model

```python
# Used Agglomerative Clustering Method
from sklearn.cluster import AgglomerativeClustering
# Create model object
model = AgglomerativeClustering(n_clusters=2, linkage='ward') # determind devide custer 2
# Fit(Train) and predict model
pred = model.fit_predict(data)
```

```python
# Show Predict data
pred
```

**Inside pred value in data set**

```python
data['cluster'] = pred # Create new column in cluster
data
```

**Devide Cluster**

```python
cluster1 = data[data['cluster']==0] # Devide cluster 1
cluster2 = data[data['cluster']==1] # Devide cluster 2

# Use Plot
plt.scatter(cluster1.x, cluster1.y, color='red') # Plot cluster 1
plt.scatter(cluster2.x, cluster2.y, color='blue') # Plot cluster 2
```
