**X value (Should be 2d array)**

```python
x = data[['age']]
x
```

**Y value (Not Should be 2d array)**

```python
y = data[['job']]
y
```

**Train and Test Values**

* Use and Import train_test_split method

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

* x train values

```python
x_train
```

* Num of x train values

```python
len(x_train)
```

* Num of x test values

```python
len(x_test)
```

**Make Model**

* Import and Use LogisticRegression Class

```python
from sklearn.linear_model import LogisticRegression
```

* Create Object

```python
model = LogisticRegression()
```

* Model train usint fit function ( fit(x,y) )

```python
model.fit(x_train,y_train)
```

**Predict Values**

* Predict x_test values

```python
model.predict(x_test)
```

```python
x_test
```

```python
y_test
```

* predict x_values and y_tests value are equal (becouse accurancy 100%)

**Accurancy**

* Use to score(x,y) Method

```python
model.score(x_test,y_test)
```

* Becouse equal to predict x_test values and normal y_test values (score is 1)
