## x data values

```python
x = data.iloc[:,1:5]
x.head()
```

###  y data values


```python
y = data.iloc[:,-1] # -1 --> Last Column
y.head()
```

### x values is Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
x[0:5] # Show between 1 to 5 values
```

### Devide Train and Test in data set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape # Show x_train data shape
```

```python
x_test.shape # Show x_train data shape
```

## Model Train Use K-Nearest Neighbors Algorithm

```python
# import KNeighborsClassifier using sklearn.neighbors class
from sklearn.neighbors import KNeighborsClassifier
#Create obgect using  KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1) # n_neighbors is K value
model.fit(x_train,y_train) # Train model using x_train,y_train data sets
```

### Using Model

```python
pred = model.predict(x_test) # Predict using x_test data set
pred[0:5] # Get predict first 5 value (Predict values)
```

```python
y_test[0:5] # Get y_test first 5 value (Actual Value)
```

### Look at Model Accurancy

**using sklearn.metricsclass**

```python
# import ccuracy_score using sklearn.metricsclass
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,pred) # Give Y accual value and Model predict value
accuracy
```

```python
# import confusion_matrix using sklearn.metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,pred) # Give Y accual value and Model predict value
cm
```

### Accurancy Details

```python
# Look at Accual value and Predict value in Y
result = pd.DataFrame(data=[y_test.values,pred],index=['Y_test','Pred'])
result.transpose()
```

### Choose Currect K value

```python
correct_sum = [] # make list in currect_sum
for i in range(1,21): # Seacrch for K value between 1 to 20
  model = KNeighborsClassifier(n_neighbors=i) # create Object  KNeighborsClassifie and K value = (n_neighbors=i
  model.fit(x_train,y_train) # Train Model
  pred = model.predict(x_test)
  correct = np.sum(pred == y_test) # if predict value eqlual to accual value
  correct_sum.append(correct) # Sum Correct
```

```python
correct_sum
```

**Using Data Frame**

```python
result = pd.DataFrame(data=correct_sum)
result.index = result.index + 1
result
```
